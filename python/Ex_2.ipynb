```python
#EX 2 
def recaman(max_terms):
    exist = set()
    seq=list()
    n = 0 
    a = 0
    while n < max_terms:
        diff=a-n
        if diff > 0 and diff not in exist:
            a = diff
        else:
            a = a + n
        exist.add(a)
        seq.append(a)
        n += 1
    return seq
```

```python
dimension=recaman(20)

plt.figure(figsize = (18, 10))


for a in dimension:
    a1=dimension[dimension.index(a)-1]    
    r = abs((a1-a)/2)
    
    if a%2==0:
        angles = np.linspace(0 * np.pi, 1 * np.pi, 100 )
        if a1>a:
            xs = (r * np.cos(angles))+(a1-r)
            ys = r * np.sin(angles)
            plt.plot((xs) , ys, color = 'green')
        else:
            xs = (r * np.cos(angles))+(a-r)
            ys = r * np.sin(angles)
            plt.plot((xs) , ys, color = 'green')
    else:
        angles = np.linspace( np.pi, 2 * np.pi, 100 )
        if a1>a:
            xs = (r * np.cos(angles))+(a1-r)
            ys = r * np.sin(angles)
            plt.plot((xs) , ys, color = 'green')
        else:
            xs = (r * np.cos(angles))+(a-r)
            ys = r * np.sin(angles)
            plt.plot((xs) , ys, color = 'green')
```
